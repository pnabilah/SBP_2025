```python
# without deep copy
a = [1, 2, [3, 4]]
b = a      # this doesn’t copy, it just points to the same list!
b[2][0] = 99   # change inside nested list
print(a)   # → [1, 2, [99, 4]]  ← a got changed too!
print(b)

# with deep copy
import copy
c = [1, 2, [3, 4]]
d = copy.deepcopy(c)      # this doesn’t copy, it just points to the same list!
d[2][0] = 99   # change inside nested list
print(c)   # → [1, 2, [99, 4]]  ← a got changed too!
print(d)
```

```python
import copy

# pretend we have two "layers"
layers = [{'W': 1}, {'W': 2}]

# save deep copies
best_weights = [copy.deepcopy(layer) for layer in layers]

# modify the original
layers[0]['W'] = 999

print("Original:", layers)
print("Saved best:", best_weights)

# restore best
for layer, best_state in zip(layers, best_weights):
    layer.update(best_state)

print("Restored:", layers)
```
